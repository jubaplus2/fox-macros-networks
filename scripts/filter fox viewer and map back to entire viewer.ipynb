```python
viewer=pd.read_csv("../outputs/three_month_viewer_and_TV.csv", usecols=["Network", "ProgramName", "HHID", "PersonID", "PersonIntab", "ViewDate", "ViewStartTime2", "ViewTime"])
viewer
```

```python
# filter fox news viewers
fox_viewer=viewer[["Network", "HHID", "PersonID", "ViewTime"]]
fox_viewer=fox_viewer[fox_viewer["Network"].str.contains('FOXNC', regex=True)]
fox_viewer
```

```python
# group by Network, HHID, PersonID and get the accumulated view time
# filiter accumulated view time>100 mins
fox_viewtime=fox_viewer.groupby(["Network", "HHID", "PersonID"], as_index=False)["ViewTime"].sum()
fox_viewtime=fox_viewtime[fox_viewtime["ViewTime"]>=100]
fox_viewtime.rename(columns = {"ViewTime": "Accumulated_ViewTime"}, inplace = True)
fox_viewtime
```

```python
# map >100 mins of fox viewer back to all viewer
final_viewer=viewer.merge(fox_viewtime, on=["HHID", "PersonID"])
final_viewer=final_viewer[final_viewer[ViewTime]>0]
final_viewer
```

```python
final_viewer['DayParts']=np.where(final_viewer['ViewStartTime2']<1200,"1morning",
                                                                  np.where(final_viewer['ViewStartTime2']<1600,"2day",
                                                                           np.where(final_viewer['ViewStartTime2']<1900,"3earlyfringe",
                                                                                    np.where(final_viewer['ViewStartTime2']<2400,"4prime",
                                                                                             np.where(final_viewer['ViewStartTime2']<2700,"5latenight","6other")))))
final_viewer
```

```python
final_viewer['DOW']=pd.to_datetime(final_viewer['ViewDate']).dt.day_name()
final_viewer
```

```python
final_viewer.to_csv("../outputs/final_viewer_1.csv", index=False)
```

## summary part


```python
import pandas as pd
import numpy as np
import datetime

final_viewer_2=pd.read_csv("../outputs/final_viewer.csv")
final_viewer_2["HHID_PersonID"]=final_viewer_2["HHID"].astype(str)+"_"+final_viewer_2["PersonID"].astype(str)
final_viewer_2
```

```python
#Networks with the counts of Fox Viewer IDs, accumulated mins, mins per ID
final_11=final_viewer_2.groupby(["Network_x"], as_index=False)["HHID_PersonID"].agg(["nunique"]).sort_values(["nunique"], ascending=False)
final_12=final_viewer_2.groupby(["Network_x"], as_index=False)["ViewTime"].agg(["sum"]).sort_values(["sum"], ascending=False)
final_1=final_11.join(final_12)
final_1=final_1.rename(columns={"nunique": "ID counts", "sum": "mins"})
final_1["mins per ID"]= final_1["mins"]/final_1["ID counts"]
final_1=final_1.sort_values(by='mins per ID', ascending=False)
final_1
```

```python
#Dayparts with the counts of Fox Viewer IDs, accumulated mins, mins per ID
final_21=final_viewer_2.groupby(["DayParts"], as_index=False)["HHID_PersonID"].agg(["nunique"]).sort_values(["nunique"], ascending=False)
final_22=final_viewer_2.groupby(["DayParts"], as_index=False)["ViewTime"].agg(["sum"]).sort_values(["sum"], ascending=False)
final_2=final_21.join(final_22)
final_2=final_2.rename(columns={"nunique": "ID counts", "sum": "mins"})
final_2
```

```python
# Shows with the counts of Fox Viewer IDs, accumulated mins, mins per ID
final_31=final_viewer_2.groupby(["ProgramName"], as_index=False)["HHID_PersonID"].agg(["nunique"]).sort_values(["nunique"], ascending=False)
final_32=final_viewer_2.groupby(["ProgramName"], as_index=False)["ViewTime"].agg(["sum"]).sort_values(["sum"], ascending=False)
final_3=final_31.join(final_32)
final_3=final_3.rename(columns={"nunique": "ID counts", "sum": "mins"})
final_3
```

```python
# Day of week with the counts of Fox Viewer IDs, accumulated mins, mins per ID
final_41=final_viewer_2.groupby(["DOW"], as_index=False)["HHID_PersonID"].agg(["nunique"]).sort_values(["nunique"], ascending=False)
final_42=final_viewer_2.groupby(["DOW"], as_index=False)["ViewTime"].agg(["sum"]).sort_values(["sum"], ascending=False)
final_4=final_41.join(final_42)
final_4=final_4.rename(columns={"nunique": "ID counts", "sum": "mins"})
final_4
```

```python
writer = pd.ExcelWriter('../outputs/summary.xlsx', engine='xlsxwriter')
final_1.to_excel(writer, sheet_name='Network')
final_2.to_excel(writer, sheet_name='DayParts')
final_3.to_excel(writer, sheet_name='Show')
final_4.to_excel(writer, sheet_name='DOW')
writer.save()
```
